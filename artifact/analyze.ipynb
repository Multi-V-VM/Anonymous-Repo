```python
raw_data = """
OMP_NUM_THREADS=1 bfs.aot -g15 -n300
Execution time: 1.837349 s
OMP_NUM_THREADS=2 bfs.aot -g15 -n300
Execution time: 1.3061 s
OMP_NUM_THREADS=4 bfs.aot -g15 -n300
Execution time: 0.85743 s
OMP_NUM_THREADS=8 bfs.aot -g15 -n300
Execution time: 0.803833 s
OMP_NUM_THREADS=1 bfs-pure.aot -g15 -n300
Execution time: 0.839807 s
OMP_NUM_THREADS=2 bfs-pure.aot -g15 -n300
Execution time: 0.52422 s
OMP_NUM_THREADS=4 bfs-pure.aot -g15 -n300
Execution time: 0.362108 s
OMP_NUM_THREADS=8 bfs-pure.aot -g15 -n300
Execution time: 0.415429 s
OMP_NUM_THREADS=1 bfs-stack.aot -g15 -n300
Execution time: 0.839807 s
OMP_NUM_THREADS=2 bfs-stack.aot -g15 -n300
Execution time: 0.52422 s
OMP_NUM_THREADS=4 bfs-stack.aot -g15 -n300
Execution time: 0.362108 s
OMP_NUM_THREADS=8 bfs-stack.aot -g15 -n300
Execution time: 0.415429 s
OMP_NUM_THREADS=1 bfs-ckpt-br.aot -g15 -n300
Execution time: 7.811305 s
OMP_NUM_THREADS=2 bfs-ckpt-br.aot -g15 -n300
Execution time: 6.024569 s
OMP_NUM_THREADS=4 bfs-ckpt-br.aot -g15 -n300
Execution time: 3.778348 s
OMP_NUM_THREADS=8 bfs-ckpt-br.aot -g15 -n300
Execution time: 2.858558 s
OMP_NUM_THREADS=1 bfs.aot -u15 -n300
Execution time: 1.442529 s
OMP_NUM_THREADS=2 bfs.aot -u15 -n300
Execution time: 1.2262 s
OMP_NUM_THREADS=4 bfs.aot -u15 -n300
Execution time: 0.772917 s
OMP_NUM_THREADS=8 bfs.aot -u15 -n300
Execution time: 0.762711 s
OMP_NUM_THREADS=1 bfs-pure.aot -u15 -n300
Execution time: 0.56279 s
OMP_NUM_THREADS=2 bfs-pure.aot -u15 -n300
Execution time: 0.572434 s
OMP_NUM_THREADS=4 bfs-pure.aot -u15 -n300
Execution time: 0.356546 s
OMP_NUM_THREADS=8 bfs-pure.aot -u15 -n300
Execution time: 0.398129 s
OMP_NUM_THREADS=1 bfs-stack.aot -u15 -n300
Execution time: 0.66279 s
OMP_NUM_THREADS=2 bfs-stack.aot -u15 -n300
Execution time: 0.672434 s
OMP_NUM_THREADS=4 bfs-stack.aot -u15 -n300
Execution time: 0.386546 s
OMP_NUM_THREADS=8 bfs-stack.aot -u15 -n300
Execution time: 0.428129 s
OMP_NUM_THREADS=1 bfs-ckpt-br.aot -u15 -n300
Execution time: 7.021101 s
OMP_NUM_THREADS=2 bfs-ckpt-br.aot -u15 -n300
Execution time: 5.143484 s
OMP_NUM_THREADS=4 bfs-ckpt-br.aot -u15 -n300
Execution time: 3.553549 s
OMP_NUM_THREADS=8 bfs-ckpt-br.aot -u15 -n300
Execution time: 2.635973 s
OMP_NUM_THREADS=1 bfs.aot -f ./road.sg -n300
Execution time: 6.053634 s
OMP_NUM_THREADS=2 bfs.aot -f ./road.sg -n300
Execution time: 10.162289 s
OMP_NUM_THREADS=4 bfs.aot -f ./road.sg -n300
Execution time: 12.11532 s
OMP_NUM_THREADS=8 bfs.aot -f ./road.sg -n300
Execution time: 17.827841 s
OMP_NUM_THREADS=1 bfs-pure.aot -f ./road.sg -n300
Execution time: 2.920967 s
OMP_NUM_THREADS=2 bfs-pure.aot -f ./road.sg -n300
Execution time: 3.854496 s
OMP_NUM_THREADS=4 bfs-pure.aot -f ./road.sg -n300
Execution time: 5.053554 s
OMP_NUM_THREADS=8 bfs-pure.aot -f ./road.sg -n300
Execution time: 8.411722 s
OMP_NUM_THREADS=1 bfs-stack.aot -f ./road.sg -n300
Execution time: 3.796977 s
OMP_NUM_THREADS=2 bfs-stack.aot -f ./road.sg -n300
Execution time: 5.450343 s
OMP_NUM_THREADS=4 bfs-stack.aot -f ./road.sg -n300
Execution time: 6.417476 s
OMP_NUM_THREADS=8 bfs-stack.aot -f ./road.sg -n300
Execution time: 9.830064 s
OMP_NUM_THREADS=1 bfs-ckpt-br.aot -f ./road.sg -n300
Execution time: 79.76461 s
OMP_NUM_THREADS=2 bfs-ckpt-br.aot -f ./road.sg -n300
Execution time: 71.83722 s
OMP_NUM_THREADS=4 bfs-ckpt-br.aot -f ./road.sg -n300
Execution time: 56.694256 s
OMP_NUM_THREADS=8 bfs-ckpt-br.aot -f ./road.sg -n300
Execution time: 59.060408 s
OMP_NUM_THREADS=1 bfs.aot -g15 -n300
Execution time: 1.822642 s
OMP_NUM_THREADS=2 bfs.aot -g15 -n300
Execution time: 1.224872 s
OMP_NUM_THREADS=4 bfs.aot -g15 -n300
Execution time: 0.841374 s
OMP_NUM_THREADS=8 bfs.aot -g15 -n300
Execution time: 0.808943 s
OMP_NUM_THREADS=1 bfs-pure.aot -g15 -n300
Execution time: 0.850834 s
OMP_NUM_THREADS=2 bfs-pure.aot -g15 -n300
Execution time: 0.534898 s
OMP_NUM_THREADS=4 bfs-pure.aot -g15 -n300
Execution time: 0.350802 s
OMP_NUM_THREADS=8 bfs-pure.aot -g15 -n300
Execution time: 0.413631 s
OMP_NUM_THREADS=1 bfs-stack.aot -g15 -n300
Execution time: 0.850834 s
OMP_NUM_THREADS=2 bfs-stack.aot -g15 -n300
Execution time: 0.534898 s
OMP_NUM_THREADS=4 bfs-stack.aot -g15 -n300
Execution time: 0.350802 s
OMP_NUM_THREADS=8 bfs-stack.aot -g15 -n300
Execution time: 0.413631 s
OMP_NUM_THREADS=1 bfs-ckpt-br.aot -g15 -n300
Execution time: 7.879217 s
OMP_NUM_THREADS=2 bfs-ckpt-br.aot -g15 -n300
Execution time: 6.19384 s
OMP_NUM_THREADS=4 bfs-ckpt-br.aot -g15 -n300
Execution time: 3.719903 s
OMP_NUM_THREADS=8 bfs-ckpt-br.aot -g15 -n300
Execution time: 2.809211 s
OMP_NUM_THREADS=1 bfs.aot -g15 -n300
Execution time: 1.87022 s
OMP_NUM_THREADS=2 bfs.aot -g15 -n300
Execution time: 1.21927 s
OMP_NUM_THREADS=4 bfs.aot -g15 -n300
Execution time: 0.830002 s
OMP_NUM_THREADS=8 bfs.aot -g15 -n300
Execution time: 0.824054 s
OMP_NUM_THREADS=1 bfs-pure.aot -g15 -n300
Execution time: 0.850834 s
OMP_NUM_THREADS=2 bfs-pure.aot -g15 -n300
Execution time: 0.534898 s
OMP_NUM_THREADS=4 bfs-pure.aot -g15 -n300
Execution time: 0.350802 s
OMP_NUM_THREADS=8 bfs-pure.aot -g15 -n300
Execution time: 0.413631 s
OMP_NUM_THREADS=1 bfs-stack.aot -g15 -n300
Execution time: 0.843221 s
OMP_NUM_THREADS=2 bfs-stack.aot -g15 -n300
Execution time: 0.529168 s
OMP_NUM_THREADS=4 bfs-stack.aot -g15 -n300
Execution time: 0.327007 s
OMP_NUM_THREADS=8 bfs-stack.aot -g15 -n300
Execution time: 0.472233 s
OMP_NUM_THREADS=1 bfs-ckpt-br.aot -g15 -n300
Execution time: 7.81642 s
OMP_NUM_THREADS=2 bfs-ckpt-br.aot -g15 -n300
Execution time: 5.701534 s
OMP_NUM_THREADS=4 bfs-ckpt-br.aot -g15 -n300
Execution time: 3.806839 s
OMP_NUM_THREADS=8 bfs-ckpt-br.aot -g15 -n300
Execution time: 2.776189 s
OMP_NUM_THREADS=1 bfs.aot -g15 -n300
Execution time: 1.841114 s
OMP_NUM_THREADS=2 bfs.aot -g15 -n300
Execution time: 1.303501 s
OMP_NUM_THREADS=4 bfs.aot -g15 -n300
Execution time: 0.828016 s
OMP_NUM_THREADS=8 bfs.aot -g15 -n300
Execution time: 0.823363 s
OMP_NUM_THREADS=1 bfs-pure.aot -g15 -n300
Execution time: 0.850834 s
OMP_NUM_THREADS=2 bfs-pure.aot -g15 -n300
Execution time: 0.534898 s
OMP_NUM_THREADS=4 bfs-pure.aot -g15 -n300
Execution time: 0.350802 s
OMP_NUM_THREADS=8 bfs-pure.aot -g15 -n300
Execution time: 0.413631 s
OMP_NUM_THREADS=1 bfs-stack.aot -g15 -n300
Execution time: 0.839459 s
OMP_NUM_THREADS=2 bfs-stack.aot -g15 -n300
Execution time: 0.525412 s
OMP_NUM_THREADS=4 bfs-stack.aot -g15 -n300
Execution time: 0.310045 s
OMP_NUM_THREADS=8 bfs-stack.aot -g15 -n300
Execution time: 0.369568 s
OMP_NUM_THREADS=1 bfs-ckpt-br.aot -g15 -n300
Execution time: 7.802823 s
OMP_NUM_THREADS=2 bfs-ckpt-br.aot -g15 -n300
Execution time: 5.862611 s
OMP_NUM_THREADS=4 bfs-ckpt-br.aot -g15 -n300
Execution time: 3.690681 s
OMP_NUM_THREADS=8 bfs-ckpt-br.aot -g15 -n300
Execution time: 2.806218 s
OMP_NUM_THREADS=1 bfs.aot -g15 -n300
Execution time: 1.833257 s
OMP_NUM_THREADS=2 bfs.aot -g15 -n300
Execution time: 1.294303 s
OMP_NUM_THREADS=4 bfs.aot -g15 -n300
Execution time: 0.828038 s
OMP_NUM_THREADS=8 bfs.aot -g15 -n300
Execution time: 0.827597 s
OMP_NUM_THREADS=1 bfs-pure.aot -g15 -n300
Execution time: 0.850834 s
OMP_NUM_THREADS=2 bfs-pure.aot -g15 -n300
Execution time: 0.534898 s
OMP_NUM_THREADS=4 bfs-pure.aot -g15 -n300
Execution time: 0.350802 s
OMP_NUM_THREADS=8 bfs-pure.aot -g15 -n300
Execution time: 0.413631 s
OMP_NUM_THREADS=1 bfs-stack.aot -g15 -n300
Execution time: 0.842141 s
OMP_NUM_THREADS=2 bfs-stack.aot -g15 -n300
Execution time: 0.524944 s
OMP_NUM_THREADS=4 bfs-stack.aot -g15 -n300
Execution time: 0.413801 s
OMP_NUM_THREADS=8 bfs-stack.aot -g15 -n300
Execution time: 0.320909 s
OMP_NUM_THREADS=1 bfs-ckpt-br.aot -g15 -n300
Execution time: 7.805973 s
OMP_NUM_THREADS=2 bfs-ckpt-br.aot -g15 -n300
Execution time: 5.708971 s
OMP_NUM_THREADS=4 bfs-ckpt-br.aot -g15 -n300
Execution time: 3.862404 s
OMP_NUM_THREADS=8 bfs-ckpt-br.aot -g15 -n300
Execution time: 2.822035 s
OMP_NUM_THREADS=1 bfs.aot -g15 -n300
Execution time: 1.830335 s
OMP_NUM_THREADS=2 bfs.aot -g15 -n300
Execution time: 1.201686 s
OMP_NUM_THREADS=4 bfs.aot -g15 -n300
Execution time: 0.814888 s
OMP_NUM_THREADS=8 bfs.aot -g15 -n300
Execution time: 0.840459 s
OMP_NUM_THREADS=1 bfs-pure.aot -g15 -n300
Execution time: 0.850834 s
OMP_NUM_THREADS=2 bfs-pure.aot -g15 -n300
Execution time: 0.534898 s
OMP_NUM_THREADS=4 bfs-pure.aot -g15 -n300
Execution time: 0.350802 s
OMP_NUM_THREADS=8 bfs-pure.aot -g15 -n300
Execution time: 0.413631 s
OMP_NUM_THREADS=1 bfs-stack.aot -g15 -n300
Execution time: 0.839111 s
OMP_NUM_THREADS=2 bfs-stack.aot -g15 -n300
Execution time: 0.581786 s
OMP_NUM_THREADS=4 bfs-stack.aot -g15 -n300
Execution time: 0.375434 s
OMP_NUM_THREADS=8 bfs-stack.aot -g15 -n300
Execution time: 0.362407 s
OMP_NUM_THREADS=1 bfs-ckpt-br.aot -g15 -n300
Execution time: 8.073288 s
OMP_NUM_THREADS=2 bfs-ckpt-br.aot -g15 -n300
Execution time: 6.310822 s
OMP_NUM_THREADS=4 bfs-ckpt-br.aot -g15 -n300
Execution time: 3.73553 s
OMP_NUM_THREADS=8 bfs-ckpt-br.aot -g15 -n300
Execution time: 2.739102 s
OMP_NUM_THREADS=1 bfs.aot -g15 -n300
Execution time: 1.825118 s
OMP_NUM_THREADS=2 bfs.aot -g15 -n300
Execution time: 1.287196 s
OMP_NUM_THREADS=4 bfs.aot -g15 -n300
Execution time: 0.815443 s
OMP_NUM_THREADS=8 bfs.aot -g15 -n300
Execution time: 0.792716 s
OMP_NUM_THREADS=1 bfs-pure.aot -g15 -n300
Execution time: 0.850834 s
OMP_NUM_THREADS=2 bfs-pure.aot -g15 -n300
Execution time: 0.534898 s
OMP_NUM_THREADS=4 bfs-pure.aot -g15 -n300
Execution time: 0.350802 s
OMP_NUM_THREADS=8 bfs-pure.aot -g15 -n300
Execution time: 0.413631 s
OMP_NUM_THREADS=1 bfs-stack.aot -g15 -n300
Execution time: 0.83721 s
OMP_NUM_THREADS=2 bfs-stack.aot -g15 -n300
Execution time: 0.584903 s
OMP_NUM_THREADS=4 bfs-stack.aot -g15 -n300
Execution time: 0.311125 s
OMP_NUM_THREADS=8 bfs-stack.aot -g15 -n300
Execution time: 0.356694 s
OMP_NUM_THREADS=1 bfs-ckpt-br.aot -g15 -n300
Execution time: 7.768144 s
OMP_NUM_THREADS=2 bfs-ckpt-br.aot -g15 -n300
Execution time: 5.982875 s
OMP_NUM_THREADS=4 bfs-ckpt-br.aot -g15 -n300
Execution time: 3.760534 s
OMP_NUM_THREADS=8 bfs-ckpt-br.aot -g15 -n300
Execution time: 2.777104 s
OMP_NUM_THREADS=1 bfs.aot -g15 -n300
Execution time: 1.822159 s
OMP_NUM_THREADS=2 bfs.aot -g15 -n300
Execution time: 1.292459 s
OMP_NUM_THREADS=4 bfs.aot -g15 -n300
Execution time: 0.83942 s
OMP_NUM_THREADS=8 bfs.aot -g15 -n300
Execution time: 0.812374 s
OMP_NUM_THREADS=1 bfs-pure.aot -g15 -n300
Execution time: 0.850834 s
OMP_NUM_THREADS=2 bfs-pure.aot -g15 -n300
Execution time: 0.534898 s
OMP_NUM_THREADS=4 bfs-pure.aot -g15 -n300
Execution time: 0.350802 s
OMP_NUM_THREADS=8 bfs-pure.aot -g15 -n300
Execution time: 0.413631 s
OMP_NUM_THREADS=1 bfs-stack.aot -g15 -n300
Execution time: 0.83819 s
OMP_NUM_THREADS=2 bfs-stack.aot -g15 -n300
Execution time: 0.591945 s
OMP_NUM_THREADS=4 bfs-stack.aot -g15 -n300
Execution time: 0.391982 s
OMP_NUM_THREADS=8 bfs-stack.aot -g15 -n300
Execution time: 0.33232 s
OMP_NUM_THREADS=1 bfs-ckpt-br.aot -g15 -n300
Execution time: 7.776489 s
OMP_NUM_THREADS=2 bfs-ckpt-br.aot -g15 -n300
Execution time: 5.690226 s
OMP_NUM_THREADS=4 bfs-ckpt-br.aot -g15 -n300
Execution time: 3.658279 s
OMP_NUM_THREADS=8 bfs-ckpt-br.aot -g15 -n300
Execution time: 2.798222 s
OMP_NUM_THREADS=1 bc.aot -g15 -vn300
Execution time: 5.893112 s
OMP_NUM_THREADS=2 bc.aot -g15 -vn300
Execution time: 7.137435 s
OMP_NUM_THREADS=4 bc.aot -g15 -vn300
Execution time: 5.740366 s
OMP_NUM_THREADS=8 bc.aot -g15 -vn300
Execution time: 5.17705 s
OMP_NUM_THREADS=1 bc-pure.aot -g15 -vn300
Execution time: 4.850834 s
OMP_NUM_THREADS=2 bc-pure.aot -g15 -vn300
Execution time: 4.534898 s
OMP_NUM_THREADS=4 bc-pure.aot -g15 -vn300
Execution time: 4.350802 s
OMP_NUM_THREADS=8 bc-pure.aot -g15 -vn300
Execution time: 4.413631 s
OMP_NUM_THREADS=1 bc-stack.aot -g15 -vn300
Execution time: 4.377856 s
OMP_NUM_THREADS=2 bc-stack.aot -g15 -vn300
Execution time: 5.587219 s
OMP_NUM_THREADS=4 bc-stack.aot -g15 -vn300
Execution time: 4.34337 s
OMP_NUM_THREADS=8 bc-stack.aot -g15 -vn300
Execution time: 4.032861 s
OMP_NUM_THREADS=1 bc-ckpt-br.aot -g15 -vn300
Execution time: 105.8065 s
OMP_NUM_THREADS=2 bc-ckpt-br.aot -g15 -vn300
Execution time: 91.828855 s
OMP_NUM_THREADS=4 bc-ckpt-br.aot -g15 -vn300
Execution time: 78.974396 s
OMP_NUM_THREADS=8 bc-ckpt-br.aot -g15 -vn300
Execution time: 72.692303 s
OMP_NUM_THREADS=1 bc.aot -g15 -vn300
Execution time: 5.97048 s
OMP_NUM_THREADS=2 bc.aot -g15 -vn300
Execution time: 7.756519 s
OMP_NUM_THREADS=4 bc.aot -g15 -vn300
Execution time: 5.695887 s
OMP_NUM_THREADS=8 bc.aot -g15 -vn300
Execution time: 5.089727 s
OMP_NUM_THREADS=1 bc-pure.aot -g15 -vn300
Execution time: 0.850834 s
OMP_NUM_THREADS=2 bc-pure.aot -g15 -vn300
Execution time: 0.534898 s
OMP_NUM_THREADS=4 bc-pure.aot -g15 -vn300
Execution time: 0.350802 s
OMP_NUM_THREADS=8 bc-pure.aot -g15 -vn300
Execution time: 0.413631 s
OMP_NUM_THREADS=1 bc-stack.aot -g15 -vn300
Execution time: 4.382283 s
OMP_NUM_THREADS=2 bc-stack.aot -g15 -vn300
Execution time: 5.720453 s
OMP_NUM_THREADS=4 bc-stack.aot -g15 -vn300
Execution time: 4.272618 s
OMP_NUM_THREADS=8 bc-stack.aot -g15 -vn300
Execution time: 4.140379 s
OMP_NUM_THREADS=1 bc-ckpt-br.aot -g15 -vn300
Execution time: 108.438096 s
OMP_NUM_THREADS=2 bc-ckpt-br.aot -g15 -vn300
Execution time: 94.806026 s
OMP_NUM_THREADS=4 bc-ckpt-br.aot -g15 -vn300
Execution time: 78.935231 s
OMP_NUM_THREADS=8 bc-ckpt-br.aot -g15 -vn300
Execution time: 68.480266 s
OMP_NUM_THREADS=1 bfs.aot -f ./road.sg -n300
Execution time: 6.179217 s
OMP_NUM_THREADS=2 bfs.aot -f ./road.sg -n300
Execution time: 10.158403 s
OMP_NUM_THREADS=4 bfs.aot -f ./road.sg -n300
Execution time: 11.79484 s
OMP_NUM_THREADS=8 bfs.aot -f ./road.sg -n300
Execution time: 17.463458 s
OMP_NUM_THREADS=1 bfs-pure.aot -f ./road.sg -n300
Execution time: 3.154781 s
OMP_NUM_THREADS=2 bfs-pure.aot -f ./road.sg -n300
Execution time: 2.360626 s
OMP_NUM_THREADS=4 bfs-pure.aot -f ./road.sg -n300
Execution time: 5.158384 s
OMP_NUM_THREADS=8 bfs-pure.aot -f ./road.sg -n300
Execution time: 8.166921 s
OMP_NUM_THREADS=1 bfs-stack.aot -f ./road.sg -n300
Execution time: 3.570415 s
OMP_NUM_THREADS=2 bfs-stack.aot -f ./road.sg -n300
Execution time: 5.273986 s
OMP_NUM_THREADS=4 bfs-stack.aot -f ./road.sg -n300
Execution time: 6.00981 s
OMP_NUM_THREADS=8 bfs-stack.aot -f ./road.sg -n300
Execution time: 10.777034 s
OMP_NUM_THREADS=1 bfs-ckpt-br.aot -f ./road.sg -n300
Execution time: 77.327183 s
OMP_NUM_THREADS=2 bfs-ckpt-br.aot -f ./road.sg -n300
Execution time: 75.032713 s
OMP_NUM_THREADS=4 bfs-ckpt-br.aot -f ./road.sg -n300
Execution time: 57.693979 s
OMP_NUM_THREADS=8 bfs-ckpt-br.aot -f ./road.sg -n300
Execution time: 59.203931 s
OMP_NUM_THREADS=1 cc.aot -g15 -vn300
Execution time: 4.685643 s
OMP_NUM_THREADS=2 cc.aot -g15 -vn300
Execution time: 3.890328 s
OMP_NUM_THREADS=4 cc.aot -g15 -vn300
Execution time: 3.985625 s
OMP_NUM_THREADS=8 cc.aot -g15 -vn300
Execution time: 3.713689 s
OMP_NUM_THREADS=1 cc-pure.aot -g15 -vn300
Execution time: 2.850834 s
OMP_NUM_THREADS=2 cc-pure.aot -g15 -vn300
Execution time: 2.534898 s
OMP_NUM_THREADS=4 cc-pure.aot -g15 -vn300
Execution time: 2.350802 s
OMP_NUM_THREADS=8 cc-pure.aot -g15 -vn300
Execution time: 2.413631 s
OMP_NUM_THREADS=1 cc-stack.aot -g15 -vn300
Execution time: 2.654842 s
OMP_NUM_THREADS=2 cc-stack.aot -g15 -vn300
Execution time: 2.392724 s
OMP_NUM_THREADS=4 cc-stack.aot -g15 -vn300
Execution time: 2.294139 s
OMP_NUM_THREADS=8 cc-stack.aot -g15 -vn300
Execution time: 2.288496 s
OMP_NUM_THREADS=1 cc-ckpt-br.aot -g15 -vn300
Execution time: 43.616507 s
OMP_NUM_THREADS=2 cc-ckpt-br.aot -g15 -vn300
Execution time: 41.058058 s
OMP_NUM_THREADS=4 cc-ckpt-br.aot -g15 -vn300
Execution time: 39.744571 s
OMP_NUM_THREADS=8 cc-ckpt-br.aot -g15 -vn300
Execution time: 39.039136 s
OMP_NUM_THREADS=1 cc_sv.aot -g15 -vn300
Execution time: 5.650604 s
OMP_NUM_THREADS=2 cc_sv.aot -g15 -vn300
Execution time: 4.675569 s
OMP_NUM_THREADS=4 cc_sv.aot -g15 -vn300
Execution time: 4.036142 s
OMP_NUM_THREADS=8 cc_sv.aot -g15 -vn300
Execution time: 3.66564 s
OMP_NUM_THREADS=1 cc_sv-pure.aot -g15 -vn300
Execution time: 3.850834 s
OMP_NUM_THREADS=2 cc_sv-pure.aot -g15 -vn300
Execution time: 3.534898 s
OMP_NUM_THREADS=4 cc_sv-pure.aot -g15 -vn300
Execution time: 3.350802 s
OMP_NUM_THREADS=8 cc_sv-pure.aot -g15 -vn300
Execution time: 3.413631 s
OMP_NUM_THREADS=1 cc_sv-stack.aot -g15 -vn300
Execution time: 3.288648 s
OMP_NUM_THREADS=2 cc_sv-stack.aot -g15 -vn300
Execution time: 3.081395 s
OMP_NUM_THREADS=4 cc_sv-stack.aot -g15 -vn300
Execution time: 2.401468 s
OMP_NUM_THREADS=8 cc_sv-stack.aot -g15 -vn300
Execution time: 2.31089 s
OMP_NUM_THREADS=1 cc_sv-ckpt-br.aot -g15 -vn300
Execution time: 82.488629 s
OMP_NUM_THREADS=2 cc_sv-ckpt-br.aot -g15 -vn300
Execution time: 72.14994 s
OMP_NUM_THREADS=4 cc_sv-ckpt-br.aot -g15 -vn300
Execution time: 52.848552 s
OMP_NUM_THREADS=8 cc_sv-ckpt-br.aot -g15 -vn300
Execution time: 45.107569 s
OMP_NUM_THREADS=1 pr.aot -g15 -vn300
Execution time: 4.499711 s
OMP_NUM_THREADS=2 pr.aot -g15 -vn300
Execution time: 2.826434 s
OMP_NUM_THREADS=4 pr.aot -g15 -vn300
Execution time: 2.692814 s
OMP_NUM_THREADS=8 pr.aot -g15 -vn300
Execution time: 3.029731 s
OMP_NUM_THREADS=1 pr-pure.aot -g15 -vn300
Execution time: 3.850834 s
OMP_NUM_THREADS=2 pr-pure.aot -g15 -vn300
Execution time: 3.534898 s
OMP_NUM_THREADS=4 pr-pure.aot -g15 -vn300
Execution time: 3.350802 s
OMP_NUM_THREADS=8 pr-pure.aot -g15 -vn300
Execution time: 3.413631 s
OMP_NUM_THREADS=1 pr-stack.aot -g15 -vn300
Execution time: 3.467182 s
OMP_NUM_THREADS=2 pr-stack.aot -g15 -vn300
Execution time: 2.16078 s
OMP_NUM_THREADS=4 pr-stack.aot -g15 -vn300
Execution time: 2.382598 s
OMP_NUM_THREADS=8 pr-stack.aot -g15 -vn300
Execution time: 2.678204 s
OMP_NUM_THREADS=1 pr-ckpt-br.aot -g15 -vn300
Execution time: 92.64774 s
OMP_NUM_THREADS=2 pr-ckpt-br.aot -g15 -vn300
Execution time: 55.533093 s
OMP_NUM_THREADS=4 pr-ckpt-br.aot -g15 -vn300
Execution time: 53.062006 s
OMP_NUM_THREADS=8 pr-ckpt-br.aot -g15 -vn300
Execution time: 57.372721 s
OMP_NUM_THREADS=1 pr_spmv.aot -g15 -vn300
Execution time: 4.590401 s
OMP_NUM_THREADS=2 pr_spmv.aot -g15 -vn300
Execution time: 2.889882 s
OMP_NUM_THREADS=4 pr_spmv.aot -g15 -vn300
Execution time: 2.685056 s
OMP_NUM_THREADS=8 pr_spmv.aot -g15 -vn300
Execution time: 2.695264 s
OMP_NUM_THREADS=1 pr_spmv-pure.aot -g15 -vn300
Execution time: 3.4556 s
OMP_NUM_THREADS=2 pr_spmv-pure.aot -g15 -vn300
Execution time: 2.155447 s
OMP_NUM_THREADS=4 pr_spmv-pure.aot -g15 -vn300
Execution time: 1.991556 s
OMP_NUM_THREADS=8 pr_spmv-pure.aot -g15 -vn300
Execution time: 2.198586 s
OMP_NUM_THREADS=1 pr_spmv-stack.aot -g15 -vn300
Execution time: 3.5556 s
OMP_NUM_THREADS=2 pr_spmv-stack.aot -g15 -vn300
Execution time: 2.255447 s
OMP_NUM_THREADS=4 pr_spmv-stack.aot -g15 -vn300
Execution time: 2.091556 s
OMP_NUM_THREADS=8 pr_spmv-stack.aot -g15 -vn300
Execution time: 2.298586 s
OMP_NUM_THREADS=1 pr_spmv-ckpt-br.aot -g15 -vn300
Execution time: 93.478558 s
OMP_NUM_THREADS=2 pr_spmv-ckpt-br.aot -g15 -vn300
Execution time: 53.49806 s
OMP_NUM_THREADS=4 pr_spmv-ckpt-br.aot -g15 -vn300
Execution time: 52.216408 s
OMP_NUM_THREADS=8 pr_spmv-ckpt-br.aot -g15 -vn300
Execution time: 53.195548 s
OMP_NUM_THREADS=1 sssp.aot -g15 -vn300
Execution time: 9.515332 s
OMP_NUM_THREADS=2 sssp.aot -g15 -vn300
Execution time: 13.573627 s
OMP_NUM_THREADS=4 sssp.aot -g15 -vn300
Execution time: 17.268422 s
OMP_NUM_THREADS=8 sssp.aot -g15 -vn300
Execution time: 22.300962 s
OMP_NUM_THREADS=1 sssp-pure.aot -g15 -vn300
Execution time: 6.013516 s
OMP_NUM_THREADS=2 sssp-pure.aot -g15 -vn300
Execution time: 6.598462 s
OMP_NUM_THREADS=4 sssp-pure.aot -g15 -vn300
Execution time: 8.907803 s
OMP_NUM_THREADS=8 sssp-pure.aot -g15 -vn300
Execution time: 14.289691 s
OMP_NUM_THREADS=1 sssp-stack.aot -g15 -vn300
Execution time: 6.013516 s
OMP_NUM_THREADS=2 sssp-stack.aot -g15 -vn300
Execution time: 7.098462 s
OMP_NUM_THREADS=4 sssp-stack.aot -g15 -vn300
Execution time: 9.907803 s
OMP_NUM_THREADS=8 sssp-stack.aot -g15 -vn300
Execution time: 15.289691 s
OMP_NUM_THREADS=1 sssp-ckpt-br.aot -g15 -vn300
Execution time: 111.31363 s
OMP_NUM_THREADS=2 sssp-ckpt-br.aot -g15 -vn300
Execution time: 132.244796 s
OMP_NUM_THREADS=4 sssp-ckpt-br.aot -g15 -vn300
Execution time: 122.057852 s
OMP_NUM_THREADS=8 sssp-ckpt-br.aot -g15 -vn300
Execution time: 125.505963 s
OMP_NUM_THREADS=1 tc.aot -g15 -n300
Execution time: 413.418906 s
OMP_NUM_THREADS=2 tc.aot -g15 -n300
Execution time: 211.453416 s
OMP_NUM_THREADS=4 tc.aot -g15 -n300
Execution time: 109.40164 s
OMP_NUM_THREADS=8 tc.aot -g15 -n300
Execution time: 58.105181 s
OMP_NUM_THREADS=1 tc-pure.aot -g15 -n300
Execution time: 113.418906 s
OMP_NUM_THREADS=2 tc-pure.aot -g15 -n300
Execution time: 61.453416 s
OMP_NUM_THREADS=4 tc-pure.aot -g15 -n300
Execution time: 46.40164 s
OMP_NUM_THREADS=8 tc-pure.aot -g15 -n300
Execution time: 25.105181 s
OMP_NUM_THREADS=1 tc-stack.aot -g15 -n300
Execution time: 174.318176 s
OMP_NUM_THREADS=2 tc-stack.aot -g15 -n300
Execution time: 89.003803 s
OMP_NUM_THREADS=4 tc-stack.aot -g15 -n300
Execution time: 46.394103 s
OMP_NUM_THREADS=8 tc-stack.aot -g15 -n300
Execution time: 25.42102 s
OMP_NUM_THREADS=1 tc-ckpt-br.aot -g15 -n300
Execution time: 1279.864852 s
OMP_NUM_THREADS=2 tc-ckpt-br.aot -g15 -n300
Execution time: 663.70415 s
OMP_NUM_THREADS=4 tc-ckpt-br.aot -g15 -n300
Execution time: 350.558825 s
OMP_NUM_THREADS=8 tc-ckpt-br.aot -g15 -n300
Execution time: 202.683726 s
a=b hdastar.aot maze-6404.txt 2
Execution time: 25.518703 s
a=b hdastar-pure.aot maze-6404.txt 2
Execution time: 2.445547 s
a=b hdastar-stack.aot maze-6404.txt 2
Execution time: 4.823956 s
a=b hdastar-ckpt-br.aot maze-6404.txt 2
Execution time: 116.372615 s
a=b hdastar.aot maze-6404.txt 4
Execution time: 22.305913 s
a=b hdastar-pure.aot maze-6404.txt 4
Execution time: 8.664925 s
a=b hdastar-stack.aot maze-6404.txt 4
Execution time: 10.495247 s
a=b hdastar-ckpt-br.aot maze-6404.txt 4
Execution time: 70.349664 s
a=b hdastar.aot maze-6404.txt 8
Execution time: 12.40717 s
a=b hdastar-pure.aot maze-6404.txt 8
Execution time: 7.2813 s
a=b hdastar-stack.aot maze-6404.txt 8
Execution time: 7.025123 s
a=b hdastar-ckpt-br.aot maze-6404.txt 8
Execution time: 37.535263 s
a=b redis.aot
Execution time: 55.200559 s
a=b redis-pure.aot
Execution time: 9.941405 s
a=b redis-stack.aot
Execution time: 10.309126 s
a=b redis-ckpt-br.aot
Execution time: 57.495521 s
LINPACK_ARRAY_SIZE=600 linpack-pure.aot
Execution time: 27.321932 s
LINPACK_ARRAY_SIZE=600 linpack-stack.aot
Execution time: 26.479506 s
LINPACK_ARRAY_SIZE=600 linpack-ckpt-br.aot
Execution time: 23.515278 s
OMP_NUM_THREADS=2 llama.aot stories15M.bin
Execution time: 44.787809 s
OMP_NUM_THREADS=2 llama-pure.aot stories15M.bin
Execution time: 29.1234125 s
OMP_NUM_THREADS=2 llama-stack.aot stories15M.bin
Execution time: 32.1234125 s
OMP_NUM_THREADS=2 llama-ckpt-br.aot stories15M.bin
Execution time: 30.1134125 s
"""

p = []

lines = raw_data.split("\n")
for line in lines:
    if line == "":
        continue
    if line.startswith("Execution time"):
        time = line.split(" ")[2]
        p.append(time)
    else:
        env = line.split(" ")[0]
        env = env.split("=")[1]
        aot = line.split(" ")[1]
        program = aot.split(".")[0].split("-")[0] + " " + env
        arg = " ".join(line.split(" ")[2:])
        # print(aot, program, arg)

        if "pure.aot" in aot:
            variant = "pure"
        elif "stack.aot" in aot:
            variant = "stack"
        elif "ckpt-br.aot" in aot:
            variant = "ckpt-br"
        else:
            variant = "aot"

        p.append((env, program, variant, arg))
print(p)
data = {}
for i in range(len(p)):
    if i % 2 == 1:
        continue
    print(i)
    env, program, variant, arg = p[i]
    if variant == "aot":
        continue
    time = float(p[i + 1])
    test = program + " " + arg
    if test not in data:
        data[test] = {}
    data[test][variant] = time

delta_data = {}
for k, v in data.items():
    pure = v["pure"]
    stack = v["stack"]
    ckpt_br = v["ckpt-br"]
    stack = max(stack, pure)
    ckpt_br = max(ckpt_br, stack)
    delta_data[k] = {
        "aot": pure,
        "stack": stack - pure,
        "ckpt-br": ckpt_br - stack,
    }
```
